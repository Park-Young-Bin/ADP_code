### change value

```python
score = pd.DataFrame({'국어':[100, 80], '수학':[75, 90],
                     '영어':[90, 95]}, index=['장화', '홍련'])
print('before')
display(score)
score.loc['홍련', '영어'] = 100 # 행, 열을 선택하여 특정 한 개 값만 변경
score['국어'] = score['국어'] - 5 # 열 전체 값을 변경
print('---------------------')
print('after')
display(score)
```

### condition

```python
# example data
students = pd.DataFrame({'name':['장화', '홍련', '콩쥐', '팥쥐', '해님', '달님'],
                        'kor':[70, 85, None, 100, None, 85],
                        'math':[65, 100, 80, 95, None, 70]})
students
```

```python
# 일반적인 조건
display(students[students['name'] == '장화'])
display(students[(students['kor']>=80) & (students['math']>=80)])
```

```python
# loc[조건문, '추가/변경할 컬럼명']: 조건에 맞게 새로운 행/열을 만들 수 있음.
students.loc[6, "name":'math'] = ['별님', 50, 60] # 조건에 맞는 위치에 새로운 행 추가
students.loc[(students['kor']>=80) & (students['math']>=70), '합격'] = 'Pass' # 조건에 맞게 '합격' 컬럼 추가 및 값 넣기
students.loc[students['합격'] != 'Pass', '합격'] = 'Fail' # 조건에 맞는 위치에 새로운 값 삽입
students
```

```python
# np.select: 하나의 컬럼에 여러 조건에 따라 다른 값을 반환
import numpy as np
condition_list = [(students['kor'] >= 90),
                 (students['kor'] >= 80) & (students['kor'] < 90),
                 (students['kor'] >= 70) & (students['kor'] < 80)]
choice_list = ['A', 'B', 'C']
students['점수'] = np.select(condition_list, choice_list, default='F') # (조건 목록, 선택목록, default=디폴트값)
students
```

### fillna

```python
# example
health = pd.DataFrame({'year':[2017, 2018, 2019, 2020, 2021, 2022],
                      'cm':[160, 162, 165, None, None, 166],
                      'kg': [53, 52, None, 50, 51, 54],
                      'eyesight':[1.2, None, 1.2, 1.2, 1.1, 0.8],
                      'sick':[None, None, None, 2, None, 1]})
health
```

```python
display(health.fillna(0)) # 모든 결측치를 0으로 대체
print('----------------------------------')
display(health.fillna(health.mean())) # 모든 결측치를 평균으로 대체
print('----------------------------------')
health['sick'] = health['sick'].fillna(0) # 열 별로 결측치 대체
health['kg'] = health['kg'].fillna(health['kg'].mean()) # 열 별로 결측치 대체
display(health)
print('----------------------------------')
health.fillna(method='pad', inplace=True) # 결측값을 이전 값으로 대체
display(health)
```

### drop_duplicates
- 중복행 제거 or 중복을 제외한 컬럼의 구성 요소 확인

```python
# 중복을 제외한 컬럼 구성 요소를 확인
print(health['cm'].drop_duplicates()) 
print('--------------------------')
print(set(health.cm))
```

```python
# 각 행을 하나의 객체로 보고 중복 제거
display(health)
print('----------------------------------')
display(health[['eyesight', 'sick']].drop_duplicates()) # 'eyesight', 'sick' 컬럼 기준으로 0, 1, 2 index가 중복
```

### join, merge

```python
product = pd.DataFrame({'code':['G1','G2','G3','G4'],
                       'name':['우유', '감자', '빵',' 치킨']})
sale = pd.DataFrame({'num':[1001,1002,1002,1003,1004],
                     'code':['G4','G3','G1','G3','G5'],
                    'cnt':[1,4,2,2,3]})
display(product)
display(sale)
display(sale.merge(product, on='code', how='inner'))
```

```python
sale.merge(product, on='code', how='outer', sort=True)
```

```python
sale.merge(product, left_on='code', right_on='code', how='left')
```

### groupby

```python
# class에 따른 각 컬럼의 평균과 중앙값
from sklearn.datasets import load_iris
IRIS = load_iris()
iris = pd.DataFrame(data=IRIS.data, columns=IRIS.feature_names)
iris['class'] = IRIS.target
iris['class'] = iris['class'].map({0:'setosa', 1:'vesicolor', 2:'virginica'})
display(iris.groupby(by='class').mean())
display(iris.groupby(by='class').median())
```

### 도수분포표

```python
# qcut
iris['petal width level'] = pd.qcut(iris['petal width (cm)'], q=3, labels=['short', 'middle', 'long'])
pd.Series(iris['petal width level']).value_counts().to_frame()
```

```python
# crosstab
pd.crosstab(iris['petal width level'], iris['class'])
```

### dataframe 문자열 데이터 변환

```python
# 인덱싱: Series.str()
landmark = pd.DataFrame({'name':['광화문','호미곶', '첨성대'], 
                         'location':['서울 종로구 사직로 161', 
                                     '경북 포항시 남구 호미곶면 대보리 150',
                                     '경북 경주시 인왕동 839-1']})
landmark['location'].str[3:6]
```

```python
# 분할
# expand=True: 분할된 문자열을 별도의 열로 확장
display(landmark['location'].str.split(" ", expand=True))
print('-------------------------------------------')
display(landmark['location'].str.split(" ", expand=False))
```

```python
# 분할된 문자열 중 특정 문자열만 선택
landmark['loc_1'] = landmark['location'].str.split(" ").str[0]
landmark
```

```python
# 탐색
print(landmark['location'].str.startswith('서울')) # startswith
print('-----------------------')
print(landmark['location'].str.endswith('1')) # endswith
print('-----------------------')
print(landmark['location'].str.contains('1')) # contains
```
